```python
def kmeans(data, K):
    
    mean_idx=[]

    number_of_elements = K
    min_random_value = 1
    max_random_value = 100000

    random.seed(42)
    
    for i in range(number_of_elements):
        r= random.randint(min_random_value, max_random_value)
        if r not in mean_idx: mean_idx.append(r)
            
    centroids = []
    
    for idx in mean_idx:
        centroids.append(data[idx])    
    
    centroids_changed = True
    labels_changed = True
    
    # Conditionals
    # centroids - old_centroids = 0 
    # labels - old_labels = 0
    
    m = 0
    
    labels = []
        
    while centroids_changed and labels_changed:
        m += 1
        
        old_labels = labels[:]
        old_centroids = centroids[:]
    
        labels = []
    
        for d in data:
            min_distance = np.linalg.norm(d-centroids[0])
            label = 0

            for i, center in enumerate(centroids):
                dist = np.linalg.norm(d-center)
                if dist < min_distance:
                    min_distance = dist
                    label = i

            labels.append(label)

        for i, center in enumerate(centroids):
            accumulated_sum = np.zeros((1, 128))
            number_of_data_points = 0
            for j, label in enumerate(labels):
                if label == i:
                    accumulated_sum += data[j]
                    number_of_data_points += 1

            centroids[i] = accumulated_sum/number_of_data_points
            
        centroids_changed = not np.allclose(np.asarray(centroids), np.asarray(old_centroids))
        labels_changed = not np.array_equal(np.asarray(labels), np.asarray(old_labels))
        
    return centroids
```

```python
def extractDescriptors(descriptor):
    image_des = []
    
    for single in descriptor:
        unwrapped = []

        for e1 in single.tolist():
            unwrapped.append(e1)

        image_des.append(unwrapped)
        
    return np.matrix.transpose(np.asarray(image_des))
```

```python
def bagOfWords(image_descriptors, centroids):
    
    labels = []
    
    for descriptor in image_descriptors:
        
        min_length = np.linalg.norm(descriptor-centroids[0])
        label = 0
        
        for i, center in enumerate(centroids):
            dist = np.linalg.norm(descriptor-center)
            if dist < min_length:
                min_length = dist
                label = i
                
        labels.append(label)
            
    bagOfWords = np.zeros(len(centroids))
    
    for ele in labels:
        bagOfWords[ele] += 1
    
    return bagOfWords    
```

```python
def axis_aligned(data_subset, data_labels, dimension, threshold):
    left_child = []
    left_labels = []
    
    right_child = []
    right_labels = []
    
    for j in range(len(data_subset)):
        if data_subset[j][dimension] > threshold:
            left_child.append(data_subset[j])
            left_labels.append(data_labels[j])
        else:
            right_child.append(data_subset[j])
            right_labels.append(data_labels[j])
            
    return right_child, right_labels, left_child, left_labels    
```

```python
def two_pixel_test(data_subset, data_labels, dimension1, dimension2, threshold):
    left_child = []
    left_labels = []
    
    right_child = []
    right_labels = []
    
    for i in range(len(data_subset)):
        if data_subset[i][dimension1]-data_subset[i][dimension2] > threshold:
            left_child.append(data_subset[i])
            left_labels.append(left_labels[i])
        else:
            right_child.append(data_subset[i])
            right_labels.append(right_labels[i])

    return right_child, right_labels, left_child, left_labels
```

```python
def shannon_entropy(data_labels):
    accumulated_class = np.zeros(10)
    entropy = 0
    
    if len(data_labels) == 0:
        return 0
    
    for i in range(len(data_labels)):
        accumulated_class[data_labels[i]] += 1

    for i in range(10):
        prob_of_class = accumulated_class[i]/len(data_labels)
        if prob_of_class > 0: entropy += prob_of_class*math.log2(prob_of_class)
#         print(len(data_labels))
#         print(entropy)

    return -1*entropy        
```

```python
def information_gain(parent_data_labels, left_labels, right_labels):
    
    if len(parent_data_labels) == 0:
        return 0
    
    parent_entropy = shannon_entropy(parent_data_labels)
    left_entropy = (len(left_labels)/len(parent_data_labels))*shannon_entropy(left_labels)
    right_entropy = (len(right_labels)/len(parent_data_labels))*shannon_entropy(right_labels)
    
    return parent_entropy - (left_entropy+right_entropy)    
```

```python
# max_depth=5, min_info_gain=0.05, min_data_points=5

def decision_tree(data_set, data_labels, opt_i_list, opt_t_list, leaf_node_stats, depth, stopping_criteria=[5, -40, 5]):
    
    opt_i = []
    opt_t = []
    
    leaf_node_stats = []
    
    obj_fun = shannon_entropy(data_labels)
    number_data_points = len(data_set)
    
    right_child, right_labels, left_child, left_labels = axis_aligned(data_set, data_labels, 0, 0)
    info_gain = information_gain(data_labels, left_labels, right_labels)
#     print(info_gain)
    
    if(depth >= stopping_criteria[0]): # or info_gain <= stopping_criteria[1] or number_data_points <= stopping_criteria[2]):
        
        depth += 1
        
        leaf_node = np.zeros(10)
        
        for i in range(len(data_set)):
            leaf_node[data_labels[i]] += 1
            leaf_node_stats.append(np.argmax(leaf_node))
            
        print("leaf " + str(depth))
        
        return opt_i, opt_t, leaf_node_stats
    
    if len(data_set) != 0:
        for i in range(len(data_set[0])):
            for t in range(255):
                right_child, right_labels, left_child, left_labels = axis_aligned(data_set, data_labels, i, t)
                info_gain = information_gain(data_labels, left_labels, right_labels)
                if info_gain < obj_fun:
                    obj_fun = info_gain
                    opt_i.append(i)
                    opt_t.append(t)

        right_child, right_labels, left_child, left_labels = axis_aligned(data_set, data_labels, opt_i[len(opt_i)-1], opt_t[len(opt_t)-1])

        depth += 1

        print("right " + str(depth))
        right_opt_i, right_opt_t, right_leaf_label  = decision_tree(right_child, right_labels, opt_i, opt_t, leaf_node_stats, depth, stopping_criteria)
        print("left " + str(depth))
        left_opt_i, left_opt_t, left_leaf_label  = decision_tree(left_child, left_labels, opt_i, opt_t, leaf_node_stats, depth, stopping_criteria)

    #     return right_opt_i, right_opt_t, right_leaf_label, left_opt_i, left_opt_t, left_leaf_label
        return right_opt_i+left_opt_i, right_opt_t+left_opt_t, right_leaf_label+left_leaf_label
    
    return opt_i, opt_t, leaf_node_stats
```
